This example uses a synthetic dataset to illustrate how to find motifs.

```python
import matrixprofile as mp
from matrixprofile.visualize import plot_motifs_mp
from matplotlib import pyplot as plt
import matplotlib

%matplotlib inline
```

```python
matplotlib.rcParams['figure.figsize'] = (15.0, 10.0)
```

```python
dataset = mp.datasets.load('motifs-discords-small.txt')
```

```python
timeseries = dataset['data']
```

```python
window_size = 32
profile = mp.compute(dataset['data'], windows=window_size)
```

```python
profile = mp.discover.motifs(profile, exclusion_zone=window_size)
```

```python
mp.visualize(profile)
plt.show()
```
