```python
import numpy as np
import matrixprofile as mp
from matrixprofile.visualize import plot_snippets
from matplotlib import pyplot as plt
import matplotlib

%matplotlib inline
```

```python
matplotlib.rcParams['figure.figsize'] = (20.0, 6.0)
```

```python
timeseries = np.loadtxt('data/InternalBleeding.txt')
```

```python
pig1_x = np.arange(0, 4440)
pig2_x = np.arange(4441, 6939)
pig1_y = timeseries[0:4440]
pig2_y = timeseries[4441:]
```

```python
plt.plot(pig1_x, pig1_y, label='Pig 1')
plt.plot(pig2_x, pig2_y, c='rosybrown', label='Pig 2')
plt.title('Arterial Blood Pressure')
plt.legend()
plt.show()
```

```python
snippet_size = 177
num_snippets = 2

# Discover Snippets
snippets = mp.discover.snippets(timeseries, snippet_size, num_snippets)
```

```python
# Print Snippet Metadata
for i in range(len(snippets)):
    print('** Snippet-' + str(i+1) + ' **')
    print('Index:', snippets[i]['index'])
    print('Fraction:', snippets[i]['fraction'])
    print()

# Visualize Snippets
plot_snippets(snippets, timeseries)
plt.show()
```
